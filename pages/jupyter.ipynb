```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('../50039bsc.csv', delimiter=';', encoding='latin-1')
df['Total'] = df.Total.apply(lambda x: float(x.replace(',', '.')))
df.head()
```

```python
df.columns
```

```python
comunidades = df['Comunidades y Ciudades Autónomas'].unique() 
comunidades
```

```python
equipamiento = df.loc[:,'Tipo de equipamiento'].unique()
equipamiento
```

```python
periodo_max = max(df.loc[:,'periodo'])
periodo_min = min(df.loc[:,'periodo'])
print(periodo_max, periodo_min)
```

```python
filtered_df = df[df['Comunidades y Ciudades Autónomas'].isin(['Andalucía','Asturias (Principado de)'])]
filtered_df = filtered_df[filtered_df['Tipo de equipamiento'].isin(['Viviendas que disponen de acceso a Internet'])]
filtered_df = filtered_df[(filtered_df['periodo'] >= 2006) & (filtered_df['periodo'] <= 2021)]
filtered_df
```

```python
grouped_df = filtered_df.groupby(['Comunidades y Ciudades Autónomas', 'Tipo de equipamiento'])

# Plot line charts for each combination
chart_data = None 
for (comunidad, equipamiento), group in grouped_df:
    if chart_data is None:
        chart_data = group[['periodo', 'Total']].set_index('periodo').rename(columns={'Total': comunidad})
    else:
        chart_data = chart_data.merge(group[['periodo', 'Total']].set_index('periodo').rename(columns={'Total': comunidad}), left_index=True, right_index=True)

print(chart_data)
```
